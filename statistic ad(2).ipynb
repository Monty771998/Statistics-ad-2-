PMF(X) = P(X = x),

P(X = x) gives the probability of X taking on that value.

2)Probability Density Function-

The PDF is used for continuous random variables, which can take on any value within a specific range. The PDF provides the probability density at each point in the range, but it does not directly give the probability of obtaining a specific value.

f(x) of a continuous random variable X is defined as the derivative of the cdf F(x):

f(x)=dF(x)/dx

where "x" represents a point in the range of X
F(x) is the cumulative distribution function of X
Consider the standard normal distribution, which is a continuous distribution with mean (μ) of 0 and standard deviation (σ) of 1.

The PDF shows the probability density at each point in the range of the standard normal distribution and hence this is continous ie.there are infinite possible values of X at any given point.

Q3: What are some examples of situations where the normal distribution might be used as a model?
Explain how the parameters of the normal distribution relate to the shape of the distribution.

Examples of Normal Distribution-
The normal distribution, often referred to as the bell curve, is a probability distribution that naturally occurs 
in many real-world phenomena. Examples include:

1)Height: The heights of people in a population tend to follow a normal distribution, with most people being of average
          height and fewer people being very tall or very short.   
2)IQ scores: IQ scores are often assumed to be normally distributed.
3)Measurement errors: Errors in measurement instruments often follow a normal distribution.
4)Stock prices: Daily returns on stocks can sometimes approximate a normal distribution.
5)Parameters of the Normal Distribution


The normal distribution is defined by two parameters:

1)Mean (μ): The center of the distribution, representing the average value. It determines the location of the peak of the bell curve.
2)Standard deviation (σ): Measures the spread of the data around the mean. A larger standard deviation indicates a wider, flatter curve, while a smaller standard deviation results in a narrower, taller curve.


Relationship between parameters and shape:

1)Mean: Shifting the mean left or right moves the entire curve along the x-axis without changing its shape.
2)Standard deviation: Increasing the standard deviation flattens the curve, while decreasing it makes the curve more peaked.

Q5: What is Bernaulli Distribution? Give an Example. What is the difference between Bernoulli
Distribution and Binomial Distribution?

Bernoulli distribution is a models a single experiment with two possible outcomes: success 
(usually assigned a value of 1) or failure (usually assigned a value of 0).
      Example: Flipping a coin once. The outcome can be either heads (success) or tails (failure).   
Parameters: Probability of success (p).

Binomial distribution is a models the number of successes in a fixed number of independent Bernoulli trials.
      Example: Flipping a coin ten times and counting the number of heads.   
Parameters: Number of trials (n) and probability of success (p). 

Key difference is  Bernoulli distribution is for a single trial, while binomial distribution is for multiple trials.

Binomial distribution is essentially the sum of multiple Bernoulli trials.

Q7: Explain uniform Distribution with an example.

Uniform distribution is a probability distribution where all possible outcomes are equally likely to occur.

Example: Rolling a fair six-sided die. Each number (1 to 6) has an equal probability of 1/6.

Often used in simulations and random number generation.

Q8: What is the z score? State the importance of the z score.

Z-score is a statistical measurement that describes a value's relationship to the mean of a group of values. 
It measures the number of standard deviations a data point is from the mean. A positive z-score indicates the data
point is above the mean, while a negative z-score signifies it's below the mean.   

Importance of z-score:
   1)Standardization: It converts data points into a standard scale, allowing comparison across different datasets.   
   2)Outlier detection: Helps identify unusual data points by analyzing z-scores.   
   3)Probability calculations: Used to calculate probabilities using the standard normal distribution table.   
   4)Statistical analysis: Essential for various statistical tests and models

Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.

The Central Limit Theorem (CLT) states that the distribution of sample means approaches a normal distribution as 
the sample size gets larger, regardless of the population distribution. In simpler terms, if you take many random 
samples from a population and calculate the mean of each sample, the distribution of those sample means will be
approximately normal.   

Significance of the Central Limit Theorem:

   1)Inference: It allows us to make inferences about a population based on sample data, even if the population itself is not normally distributed.   
   2)Hypothesis testing: Many statistical tests assume normality, and the CLT helps justify this assumption.   
   3)Confidence intervals: It's used to construct confidence intervals for population parameters.   
   4)Sampling distribution: It forms the basis for understanding the behavior of sample statistics.

Q10: State the assumptions of the Central Limit Theorem.

The Central Limit Theorem relies on a few key assumptions:

  1Independence: The observations in the sample must be independent of each other.
  2)Random Sampling: The data should be obtained through a random sampling process.   
  3)Sample Size: While the CLT generally holds for large sample sizes, a commonly accepted rule of thumb is that the sample size should be at least 30.   
  4)Finite Variance: The population from which the sample is drawn should have a finite variance.
